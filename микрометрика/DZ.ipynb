```python
def filter_rows(df, column_name, value):
    return df[df[column_name] != value]
def show_categories(df, column_name):
    d = {x : len(df[df[column_name] == x]) for x in list(df[column_name].dtype.categories)}
    d.update({'Всего' : len(df)})
    return d
```

```python
df_formatted = filter_rows(edu_mar, 'u_marst', 'ЗАТРУДНЯЮСЬ ОТВЕТИТЬ')
df_formatted = filter_rows(df_formatted, 'u_marst', 'НЕТ ОТВЕТА')
df_formatted = filter_rows(df_formatted, 'u_marst', 'ОТКАЗ ОТ ОТВЕТА')
```
